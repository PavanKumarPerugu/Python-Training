## üßÆ NumPy Basics

```python
import numpy as np

# Daily temperature readings (¬∞C) for 7 days
temps = np.array([23.5, 24.0, 22.8, 25.1, 26.3, 27.0, 24.8])
print("Temperature array:", temps)

# Creating arrays using functions
days = np.arange(1, 8)  # 1 to 7
print("Days:", days)

# Creating evenly spaced values (e.g., time samples)
time = np.linspace(0, 1, 5)
print("Time samples:", time)
```

```python
production = np.array([120, 135, 150, 160, 145])
target = np.array([130, 140, 150, 150, 150])

# Difference from target
deviation = production - target
print("Deviation:", deviation)

# Mean and standard deviation
print("Average Production:", np.mean(production))
print("Std Dev:", np.std(production))

# Conditional filtering
high_performance = production[production > 140]
print("High performing days:", high_performance)
```

```python
data = np.arange(1, 13)  # 12 values
print("Production Values:\n", data)

reshaped = data.reshape(3, 4)  # 3 lines √ó 4 days
print("Production Matrix:\n", reshaped)

# Access row 2 (2nd production line)
print("Line 2 data:", reshaped[1])

# Access element (line 3, day 4)
print("Line 3, Day 4:", reshaped[2, 3])
```

## üìä Pandas Basics

```python
import pandas as pd

# Series example
sales = pd.Series([2500, 2700, 3000, 2800], index=["Week1", "Week2", "Week3", "Week4"])
print("Weekly Sales:\n", sales)

# DataFrame example
data = {
    "Product": ["Fish Fillet", "Shrimp Pack", "Crab Legs"],
    "Quantity": [120, 90, 45],
    "Price": [250, 400, 600]
}
df = pd.DataFrame(data)
print("\nProduction Data:\n", df)
```

```python
import pandas as pd

# -------------------------------
# üåü SERIES EXAMPLE ‚Äî Weekly Sales
# -------------------------------
sales = pd.Series(
    [2500, 2700, 3000, 2800],
    index=["Week1", "Week2", "Week3", "Week4"],
    name="Weekly Sales"
)

print("üìä Weekly Sales Series:\n", sales)

# Accessing by index label
print("\nSales in Week3:", sales["Week3"])

# Slicing using index labels
print("\nSales from Week2 to Week4:\n", sales["Week2":"Week4"])

# Rename index
sales.index = ["W1", "W2", "W3", "W4"]
print("\nRenamed Index:\n", sales)

# Adding a new index label with a new value
sales["W5"] = 3100
print("\nAfter adding Week 5:\n", sales)

# -------------------------------
# üßÆ DATAFRAME EXAMPLE ‚Äî Production Data
# -------------------------------
data = {
    "Product": ["Fish Fillet", "Shrimp Pack", "Crab Legs"],
    "Quantity": [120, 90, 45],
    "Price": [250, 400, 600]
}

# Create DataFrame with custom index (e.g., Product Codes)
df = pd.DataFrame(
    data,
    index=["P001", "P002", "P003"]
)
print("\nüè≠ Production Data:\n", df)

# Give the index a name
df.index.name = "Product Code"
print("\nWith Named Index:\n", df)

# Access a specific row using index label
print("\nDetails of P002:\n", df.loc["P002"])

# Access multiple rows using index list
print("\nDetails of P001 and P003:\n", df.loc[["P001", "P003"]])

# Change index to another column (e.g., Product)
df_with_product_index = df.set_index("Product")
print("\nDataFrame with Product as Index:\n", df_with_product_index)

# Reset index back to default
df_reset = df_with_product_index.reset_index()
print("\nAfter Resetting Index:\n", df_reset)
```

```python
import pandas as pd
import numpy as np

# Create dummy data
data = {
    "Date": [
        "2024-01-01", "2024-01-02", "2024-01-03",
        "2024-01-04", "2024-01-05", "2024-01-06",
        np.nan, "2024-01-07"
    ],
    "Region": ["North", "South", "East", "West", "North", "South", "East", "West"],
    "Product": ["Fish Fillet", "Shrimp Pack", "Crab Legs", "Fish Fillet",
                "Shrimp Pack", "Crab Legs", "Fish Fillet", "Crab Legs"],
    "Quantity": [120, 90, np.nan, 140, 80, 45, 120, 45],  # Missing + duplicate values
    "Revenue": [30000, 32000, 36000, np.nan, 28000, 27000, 30000, 27000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Add a duplicate row intentionally
df = pd.concat([df, df.iloc[[2]]], ignore_index=True)

# Save to CSV
df.to_csv("sample_data.csv", index=False)

print("‚úÖ Dummy CSV file 'sample_data.csv' created successfully!")
print("\nPreview:")
print(df)
```

```python
# Load CSV (you can use any local CSV file)
df = pd.read_csv("sample_data.csv")

# View data
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Fill missing numeric values with mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Drop duplicates
df.drop_duplicates(inplace=True)

# Convert date column to datetime (if exists)
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])

print("\nCleaned Data Info:")
print(df.info())
```

```python
# Summary statistics
print(df.describe())

# Check correlations
print("\nCorrelation Matrix:\n", df.corr(numeric_only=True))

# Group by example (e.g., average revenue by region)
if 'Region' in df.columns and 'Revenue' in df.columns:
    region_summary = df.groupby('Region')['Revenue'].mean()
    print("\nAverage Revenue by Region:\n", region_summary)
```

## üìà Data Visualization (Matplotlib)

```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
production = [120, 135, 150, 160, 145]

# Line plot
plt.plot(days, production, marker='o', color='blue', label='Production')
plt.title("Daily Production Trend")
plt.xlabel("Day")
plt.ylabel("Units Produced")
plt.legend()
plt.show()

# Bar plot
plt.bar(days, production, color='skyblue')
plt.title("Production Volume per Day")
plt.show()
```

```python
# Scatter plot
temperature = [20, 22, 23, 25, 27, 28, 30]
output = [200, 210, 215, 230, 250, 260, 270]

plt.scatter(temperature, output, color='red')
plt.title("Production vs Temperature")
plt.xlabel("Temperature (¬∞C)")
plt.ylabel("Output (kg)")
plt.show()

# Histogram
plt.hist(production, bins=5, color='green', edgecolor='black')
plt.title("Production Distribution")
plt.xlabel("Units Produced")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load CSV
file = input("Enter CSV file name: ")
df = pd.read_csv(file)

# Step 2: Show basic info
print("\n--- DATA INFO ---")
print(df.info())
print("\n--- FIRST 5 ROWS ---")
print(df.head())

# Step 3: Handle missing data
print("\nMissing values before cleaning:\n", df.isnull().sum())
df.fillna(df.mean(numeric_only=True), inplace=True)

# Step 4: Summary statistics
print("\n--- SUMMARY ---")
print(df.describe())

# Step 5: Visualization
if 'Region' in df.columns and 'Revenue' in df.columns:
    region_sum = df.groupby('Region')['Revenue'].sum()
    region_sum.plot(kind='bar', color='orange', title="Total Revenue by Region")
    plt.ylabel("Revenue")
    plt.show()

# Histogram for numeric data
df.hist(figsize=(8, 6), color='skyblue', edgecolor='black')
plt.suptitle("Data Distributions")
plt.tight_layout()
plt.show()

# Step 6: Correlation analysis
print("\nCorrelation Matrix:\n", df.corr(numeric_only=True))

# Step 7: Export cleaned data
df.to_csv("cleaned_data.csv", index=False)
print("\nCleaned data saved as cleaned_data.csv ‚úÖ")
```
